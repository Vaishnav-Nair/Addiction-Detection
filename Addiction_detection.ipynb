```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping

# Encode labels
le = LabelEncoder()
df['label'] = le.fit_transform(df['addiction_type'])

# Prepare text data
X = df['text'].values
y = df['label'].values
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
sequences = tokenizer.texts_to_sequences(X)
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, y, test_size=0.2, random_state=42)

# Model creation
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))
model.add(Bidirectional(LSTM(64, return_sequences=True)))
model.add(Dropout(0.5))  # Dropout layer for regularization
model.add(LSTM(32))
model.add(Dropout(0.5))  # Dropout layer for regularization
model.add(Dense(len(le.classes_), activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, callbacks=[early_stopping])

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Accuracy: {accuracy:.2f}')
```

```python
def predict_addiction_type(input_text, model, tokenizer, label_encoder, max_length, threshold=0.5):
    # Preprocess input text
    seq = tokenizer.texts_to_sequences([input_text])
    padded_seq = pad_sequences(seq, maxlen=max_length)

    # Make prediction
    prediction = model.predict(padded_seq)
    predicted_class_index = np.argmax(prediction, axis=-1)
    confidence_score = np.max(prediction)  # Get the confidence score for the predicted class

    # Decode the predicted class index to get the actual addiction type
    predicted_addiction_type = label_encoder.inverse_transform(predicted_class_index)

    # Check confidence score and return "none" if below threshold
    if confidence_score < threshold:
        return "none", confidence_score  # Return confidence score

    return predicted_addiction_type[0], confidence_score  # Return addiction type and confidence score

```
